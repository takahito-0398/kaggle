```python
import numpy as np
import pandas as pd
```

```python
train_dir="./train_type5"
df_cm = pd.read_csv("./pred_by_traindata/pred_train_ResNet50V2_type5_3epochs.csv ")
df_cm["path"] = train_dir + "/" + df_cm["id"] + ".npy"
```

```python
from keras.utils import np_utils

def getaa():
    batch_size = 10
    da_ratio = 1.7
    n_da_per_batch = int(batch_size * (da_ratio - 1))
    df_aug = df_cm[df_cm["cm"] == "FN"].sample(n_da_per_batch)
    
    X_batch_aug = np.array([np.load(file) for file in df_aug.path.values])
    y_batch_aug = df_aug.true_label.values
    INPUT_SHAPE = (64, 129, 3)
    X_batch_aug = X_batch_aug.reshape((-1, *INPUT_SHAPE))
    y_batch_aug = np_utils.to_categorical(y_batch_aug, 2)

    return X_batch_aug, y_batch_aug
```

```python
x1, y1 = getaa()
x1.shape, y1.shape
```
